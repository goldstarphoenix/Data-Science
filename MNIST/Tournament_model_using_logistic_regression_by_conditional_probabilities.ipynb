# Defining the classifiers and training the model 






```
#use pandas to read the csv 

train_data=pd.read_csv(r'/content/gdrive/My Drive/Data Science/mnist_train.csv')
test_data=pd.read_csv(r'/content/gdrive/My Drive/Data Science/mnist_test.csv')

scale=StandardScaler() #use feature scaling 

# obtaining training and test data sets

train_label=train_data.iloc[:,0] # choosing all the rows and only the first column
train_data=train_data.iloc[:,1:] #choosing all the rows and leaving out the first column alone
test_label=test_data.iloc[:,0]
test_data=test_data.iloc[:,1:]
train_scaled=scale.fit_transform(train_data)
test_scaled=scale.fit_transform(test_data)

# initializing matrices to an empty one
score=np.empty(10)# to get an error free item assignment in numpy array
train_bclf=np.empty(train_label.shape)
test_bclf=np.empty(test_label.shape)
log_reg=list()
pred_label=list()
confusion=list()
#prediction=list()
log_reg=[] # an empty list
```

```
for k in range(10): # for train labels of each binary classifier
  for i in range(60000):
    if train_label[i]==k:
      train_bclf[i]=0 # 1 means that the binary classifier classifies the label as one of the j's 
    else:
      train_bclf[i]=1
  for j in range(10000): # for test labels of each binary classifier 
    if test_label[j]==k:
      test_bclf[j]=0 # 1 means that the binary classifier classifies the label as one of the k's 
    else:
      test_bclf[j]=1
  
  # model is trained
  from sklearn.linear_model import LogisticRegression
  log_reg.insert(k,LogisticRegression(solver = 'lbfgs')) # lbfgs is the fastest solver
                                                         # inserts at the kth index a log regressor
  
  # fit the model on training data
  log_reg[k].fit(train_scaled, train_bclf) # fitting every kth index regressor on the training data
  
```

Making predictions for every test data point

```
prediction=[]
for i in range(10000):
  prob=np.empty(10) # defining an empty array
  maxim=0 
  var=0
  for k in range(10): # for every test data a probability is assigned in this inner loop
    prob[k]=(log_reg[k].predict_proba(test_scaled[i:i+1])[:,0]) # selects the probability of identifying i as k
    maxim=max(maxim,prob[k]*100) # compares every previous maxim with the present maxim and gives the highest prob
    if maxim==prob[k]*100:
      var=k    # making the prediction as k for that particular dataset
  prediction.insert(i,var) # assigning the number that i is identified as in the ith position 
print(prediction)  
```

```
from sklearn.metrics import accuracy_score
score=accuracy_score(prediction,test_label)
print(score)
```
